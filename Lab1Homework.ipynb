```python
import pandas as pd
dulieuxettuyen = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/dulieuxettuyendaihoc.csv")
dulieuxettuyen.head(5)
```

```python
dulieuxettuyen.tail(5)
```

```python
#thay doi du lieu goc dung inplaceTrue
dulieuxettuyen['DT'].fillna(0 , inplace= True)
dulieuxettuyen["T1"].fillna(dulieuxettuyen['T1'].mean(),inplace= True)

is_float64 = dulieuxettuyen.dtypes == float
#lay ra ca kieu du leiu lac float
df_columns_list  = dulieuxettuyen.columns[is_float64]

for col in df_columns_list:
    dulieuxettuyen[col].fillna(dulieuxettuyen[col].mean(), inplace= True)

dulieuxettuyen.head(10)
```

```python
for i in ['1','2','6']:
    print("TBM"+i)

for i in ['1','2','6']:
    dulieuxettuyen["TBM"+i]=dulieuxettuyen["T"+i]*2+ dulieuxettuyen["L"+i]+ dulieuxettuyen["H"+i]+ dulieuxettuyen["S"+i]+ dulieuxettuyen["V"+i]+ dulieuxettuyen["X"+i]+ dulieuxettuyen["D"+i]+ dulieuxettuyen["N"+i]

dulieuxettuyen.rename(columns={"TBM6" : "TBM3"},inplace= True)

dulieuxettuyen.head(5)

dulieuxettuyen.to_csv("https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/dulieuxettuyendaihoc.csv")
```
