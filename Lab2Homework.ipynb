```python
import pandas as pd
import numpy as np

df = pd.read_csv("http://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/dulieuxettuyendaihoc.csv")
print(df.head(5))

```

```python
df_male = df[df['GT'] == 'M']
df_male

df_female = df[df['GT'] == 'F']
df_female

sorted_value_male = df_male.sort_values(by ='DH2')
sorted_value_female = df_female.sort_values(by = 'DH2')

print('Male')
display(sorted_value_male)
print('Female')
display(sorted_value_female)
```

Hy to pivot-table  thng k cc gi tr count, sum, mean, median, min, max, sdt, Q1,Q2 v Q3 ca DH1 theo KT

```python
print(df.groupby("KT")["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]))
df_KT=df.groupby("KT")["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]) 
print(pd.pivot_table(df_KT,columns="KT"))
```

Hy to pivot-table  thng k cc gi tr count, sum, mean, median, min, max, sdt, Q1, Q2 v Q3 ca DH1 theo KT v KV

```python
df_KT_KV = df.groupby(["KT","KV"])["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]) 
print( pd.pivot_table(df_KT_KV,columns=["KT","KV"]))
```
